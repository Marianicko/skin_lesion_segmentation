```python
# Импорт библиотек
import torch
import matplotlib.pyplot as plt
from torch.utils.tensorboard import SummaryWriter
from dataset_upd import PHDataset, get_datasets
from model import SegmentationModel
from config import Config
%matplotlib inline
```

```python
# Загрузка данных
train_dataset, val_dataset, test_dataset = get_datasets(
    images_dir="../PH2_Dataset/trainx",
    masks_dir="../PH2_Dataset/trainy"
)

# Визуализация примера
image, mask = train_dataset[6]
image = image.permute(1, 2, 0).numpy()
# Нормализуем изображение для отображения
image = (image - image.min()) / (image.max() - image.min())  # Приводим к [0, 1]

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title("Изображение")
plt.subplot(1, 2, 2)
plt.imshow(mask.squeeze().numpy(), cmap='gray')
plt.title("Маска")
plt.show()
```
